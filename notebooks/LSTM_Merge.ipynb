```python
file_map = {
    "Netflix": "NFLX_LSTM_predictions_n.csv",
    "Apple": "AAPL_LSTM_predictions_n.csv",
    "Amazon": "AMZN_LSTM_predictions_n.csv",
    "Amgen": "AMGN_LSTM_predictions_n.csv",
    "Costco": "COST_LSTM_predictions_n.csv",
    "Google": "GOOGL_LSTM_predictions_n.csv",
    "Meta": "META_LSTM_predictions_n.csv",
    "Microsoft": "MSFT_LSTM_predictions_n.csv",
    "Starbucks": "SBUX_LSTM_predictions_n.csv",
    "Tesla": "TSLA_LSTM_predictions_n.csv"
}
```

```python
import pandas as pd

merged_df = None

for i, (stock, filename) in enumerate(file_map.items()):
    df = pd.read_csv(filename, header=None)

    if i == 0:
        # Keep Date column from first file
        merged_df = pd.DataFrame()
        merged_df["Date"] = df.iloc[:, 0]

    # Add prediction column
    merged_df[f"{stock}_LSTM_LogReturns"] = df.iloc[:, 1].values

# Save final merged file
output_file = "LSTM_Merged_LogReturns_ByIndex.csv"
merged_df.to_csv(output_file, index=False)

print(f"✅ LSTM merged file saved as {output_file}")
files.download(output_file)
```
