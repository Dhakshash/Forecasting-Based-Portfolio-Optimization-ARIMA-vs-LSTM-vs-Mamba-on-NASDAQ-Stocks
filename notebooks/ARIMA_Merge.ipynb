```python
import pandas as pd
from google.colab import files

# ⬆️ Step 1: Upload all individual fo  recast CSVs from local machine
uploaded = files.upload()  # Upload all *_forecast.csv files manually

# 🧩 Step 2: Merge all into one DataFrame by Date
merged_df = None

for file_name in uploaded.keys():
    df = pd.read_csv(file_name, parse_dates=["Date"], index_col="Date")

    if merged_df is None:
        merged_df = df
    else:
        merged_df = merged_df.join(df, how='outer')  # Align by Date

# 💾 Step 3: Save and download the merged forecast
merged_df.to_csv("merged_arima_forecasts.csv")
print("✅ Merged forecast saved as 'merged_arima_forecasts.csv'")
files.download("merged_arima_forecasts.csv")
```

```python
import pandas as pd

# Load ARIMA file
arima_file = "ARIMA_Merged_LogReturns_Final (1).csv"  # Adjust path
arima_df = pd.read_csv(arima_file)

# Manual mapping from ticker to full company name
ticker_to_name = {
    "AAPL_forecast": "Apple",
    "AMZN_forecast": "Amazon",
    "AMGN_forecast": "Amgen",
    "COST_forecast": "Costco",
    "GOOGL_forecast": "Google",
    "META_forecast": "Meta",
    "MSFT_forecast": "Microsoft",
    "NFLX_forecast": "Netflix",
    "SBUX_forecast": "Starbucks",
    "TSLA_forecast": "Tesla"
}

# Rename columns
new_columns = []
for col in arima_df.columns:
    if col == "Date":
        new_columns.append("Date")
    else:
        # Map ticker to full name if exists, else keep as is
        stock_name = ticker_to_name.get(col, col)
        new_columns.append(f"{stock_name}_ARIMA_LogReturns")

arima_df.columns = new_columns

# Save cleaned ARIMA file
output_file = "ARIMA_Merged_LogReturns_Clean.csv"
arima_df.to_csv(output_file, index=False)

print(f"✅ ARIMA file cleaned and saved as {output_file}")
from google.colab import files
files.download(output_file)
```
